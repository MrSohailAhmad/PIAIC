```python
image_path = "bruchers.png"
image = PIL.Image.open(image_path)
image
```

```python
prompt = f"""
explane me about this image:

"""
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
response = model.generate_content([prompt, image])
print(response.text)
```

```python
# Display the response as Markdown
display(Markdown(f"**Gemini Response:**\n\n{response.text}"))
```
